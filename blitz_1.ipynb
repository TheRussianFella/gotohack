```python
users_scores = valuable_activity.user_id.value_counts()
```

```python
solvers = users_scores[users_scores > 23]
solvers_ids = solvers.to_dict().keys()
len(solvers_ids)
```

```python
actions_of_solvers = valuable_activity[valuable_activity.user_id.isin(solvers_ids)]
len(actions_of_solvers)
```

```python
actions_of_specific_solvers = [actions_of_solvers[actions_of_solvers.user_id == x] for x in solvers_ids]
len(actions_of_specific_solvers)
```

```python
first_actions_of_solvers = activity_data[activity_data.user_id.isin(solvers_ids)]
first_actions_of_specific_solvers = [first_actions_of_solvers[first_actions_of_solvers.user_id == x] for x in solvers_ids]
len(first_actions_of_specific_solvers)
```

```python
reach_of_24 = [action_of_solver.time.tolist()[len(action_of_solver.time.tolist()) - 24] for action_of_solver in actions_of_specific_solvers]
beggining = [action_of_solver.time.tolist()[len(action_of_solver.time.tolist()) - 1] for action_of_solver in first_actions_of_specific_solvers]
```

```python
ids = [actions_of_specific_solvers[x].user_id for x in range(len(actions_of_specific_solvers))]
ids = [user.max() for user in ids]
time_of_completion = dict((ids[x], reach_of_24[x] - beggining[x]) for x in range(len(actions_of_specific_solvers)))
```

```python
times = sorted(time_of_completion.values())[:10]
winners = [list(time_of_completion.keys())[list(time_of_completion.values()).index(x)] for x in times]
winners
```
