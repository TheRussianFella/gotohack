```python
import pandas as pd
from tqdm import tqdm_notebook
import numpy as np
```

```python
course_structure = pd.read_csv("blitz/course-217-structure.csv")
activity_data = pd.read_csv("blitz/course-217-events.csv")
```

```python
course_structure
```

```python
o = course_structure.sort_values(['module_position', 'lesson_position', 'step_position'])
order = np.array(o.step_id)
```

```python
activity_data
```

```python
user_ids = activity_data.user_id.unique()
```

```python
users_actions = {}
for user in user_ids:
    users_actions[user] = activity_data[activity_data.user_id == user][['step_id', 'time']]
    users_actions[user].reset_index(drop=True)
```

```python
reverse_order = {}
for key, value in enumerate(order):
    reverse_order[value] = key
```

```python
def is_come_back(user_id, step_id):
    next_step = order[reverse_order[step_id] + 1]
    needed_events = users_actions[user_id][users_actions[user_id].step_id.isin([step_id, next_step])]

    first_visit = needed_events[needed_events.step_id == step_id].time.min()
    first_visit_next_step = needed_events[(needed_events.step_id == next_step)
                                   & (needed_events.time > first_visit)].time.min()
    return ((needed_events.time > first_visit_next_step) & (needed_events.step_id == step_id)).any()
```

```python
def count_come_back_ratio(step_id):
    if reverse_order[step_id] >= len(order)-1:
        return 0
    
    returned = 0
    good_users = activity_data[activity_data.step_id == step_id].user_id.unique()
    
    for user_id in good_users:
        returned += is_come_back(user_id, step_id)
    
    return float(returned) / float(good_users.shape[0])
```

```python
come_back_ratio = {}
for step in tqdm_notebook(order):
    come_back_ratio[step] = count_come_back_ratio(step)
```

```python
reverse_come_back_ratio = {}
for key in come_back_ratio.keys():
    reverse_come_back_ratio[come_back_ratio[key]] = key
```

```python
top_10 = []
for value in sorted(come_back_ratio.values(), reverse=True):
    top_10.append(reverse_come_back_ratio[value])
```
